```python
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact
from IPython.display import display, clear_output

# Simulating a datacube with random values (replace with your actual data)
N = 50  # Number of images
id = 0  # Shape (N, 128, 128)

def show_image(index):
    #plt.gcf().clear()
    fig, axs = plt.subplots(1,3, figsize=(16,4))
    cbar0 = axs[0].imshow(emu[id][index], cmap='inferno')  # Change cmap if needed
    plt.colorbar(cbar0)
    axs[0].set_title(f"Emulation")
    axs[0].axis("off")
    cbar1 = axs[1].imshow(image[id][index], cmap='inferno')
    plt.colorbar(cbar1)
    axs[1].axis('off')
    axs[1].set_title('Simulation')
    axs[2].set_title('Residuals')
    cbar2 = axs[2].imshow(emu[id][index]-image[id][index], vmin=-10., vmax=10., cmap='bwr')
    axs[2].axis('off')
    plt.colorbar(cbar2)
    plt.show()

# Creating an interactive slider
clear_output(wait=True)
interact(show_image, index=(0, emu.shape[1]));
```
