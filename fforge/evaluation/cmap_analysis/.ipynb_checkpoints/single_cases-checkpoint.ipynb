```python
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact
from IPython.display import display, clear_output

# Simulating a datacube with random values (replace with your actual data)
N = 50  # Number of images
id = 0  # Shape (N, 128, 128)

def show_image(index):
    #plt.gcf().clear()
    fig, axs = plt.subplots(1,3, figsize=(16,4))
    cbar0 = axs[0].imshow(emu[id][index], cmap='inferno')  # Change cmap if needed
    plt.colorbar(cbar0)
    axs[0].set_title(f"Emulation")
    axs[0].axis("off")
    cbar1 = axs[1].imshow(simu[id][index], cmap='inferno')
    plt.colorbar(cbar1)
    axs[1].axis('off')
    axs[1].set_title('Simulation')
    axs[2].set_title('Residuals')
    cbar2 = axs[2].imshow(emu[id][index]-simu[id][index], vmin=-30., vmax=30., cmap='bwr')
    axs[2].axis('off')
    plt.colorbar(cbar2)
    plt.show()

# Creating an interactive slider
clear_output(wait=True)
interact(show_image, index=(0, emu.shape[1]));
```

```python
xy = np.linspace(-3,3,256)
xx, yy = np.meshgrid(xy, xy)
rr = np.sqrt(xx**2 + yy**2)
plt.scatter(xy[128:], emu_vx[1,0,128, 128:]*1e-2, label='emulation', marker='.')
plt.scatter(xy[128:],simu_tri[1,1,128,128:]-rr[128, 128:]**-0.5, label='simulation', marker='.')
plt.xlim(0.5,2.8)
plt.xlabel('r')
plt.ylabel('$v_\phi$')
plt.ylim(-0.3, 0.3)
plt.legend()
#plt.plot(rr[128, 128:]**-0.5)
```
